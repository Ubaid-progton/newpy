```python
import requests
from bs4 import BeautifulSoup

q = []
for x in range(1,11):
    
    url = f'https://quotes.toscrape.com/page/{x}'
    r = requests.get(url)
    soup = BeautifulSoup(r.content,"lxml")
    col = soup.find_all('div',class_="quote")
    print(url)

    for i in col:
        b = []
        div = i.find('span',class_="text").text
        tag = i.find('div',class_="tags").meta['content']
        link = i.find('a')['href']
        b.append([div])
        b.append([tag])
        b.append([link])
        print(b)
        
        
    


# for i in div:
#     b = []
#     quote = i.find('span',class_="text").text.strip()
#     tag = i.find('div',class_="tags").meta['content'].a['href']
#     b.append([quote,tag])
#     print(b)
```
