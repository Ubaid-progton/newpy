```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
f = []
for x in range(0,4):
    url = f"https://nutrition.org/business-directory/page/{x}"
    

    def ds(url_name):
        r = requests.get(url_name)
        return BeautifulSoup(r.content,"lxml")
    soup = ds(url)
    # print(url)
    supp = soup.find_all("h2",{"class":"listing-title"})
    for i in supp:
        name = i.text.strip()
        link = i.find('a')['href']
        s2 = ds(link)
        about =s2.find('div',{"class":"listing-content"}).text
        link2 =s2.find('div',{"class":"listing-url"}).a['href']
        bv = s2.find('div',{"id":"listing-overview"}).text
        f.append([name,link,about,link2,bv])
      
# data = pd.DataFrame(f,columns=["Company","link","About Prod","Prod link","Business Overview"])
# print(data)
# data.to_csv("a.csv")
        
```
